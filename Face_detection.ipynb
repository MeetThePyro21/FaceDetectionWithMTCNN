```python
# face detection with mtcnn
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from mtcnn.mtcnn import MTCNN
```

```python
# draw an image with detected objects
def draw_image_with_boxes(image_name, result_list):
    # load the image
    data = pyplot.imread(image_name)
    # plot the image
    pyplot.imshow(data)
    # get the context for drawing boxes
    ax = pyplot.gca()
    # plot each box
    for result in result_list:
        # get coordinates
        x, y, width, height = result['box']
        # create the shape
        rect = Rectangle((x, y), width, height, fill=False, color='green')
        # draw the box
        ax.add_patch(rect)
        # draw the dots
        for key, value in result['keypoints'].items():
            # create and draw dot
            dot = Circle(value, radius=5, color='red')
            ax.add_patch(dot)
    pyplot.show()
```

```python
image_name = '/Users/awesome/Downloads/me.jpg'
```

```python
# load image
pixels = pyplot.imread(image_name)
```

```python
# create the detector
detector = MTCNN()
```

```python
# detect faces in the image
faces = detector.detect_faces(pixels)
```

```python
# display face/faces
draw_image_with_boxes(image_name, faces)
```
